```python
import pandas as pd

file_path = r'c:\ai_x\source\JikFam\data\trade_weekly_galic.csv'
df = pd.read_csv(file_path, encoding='cp949')

climate_cols = [
    '일평균기온',
    '최고기온',
    '최저기온',
    '평균상대습도',
    '강수량(mm)',
    '1시간최고강수량(mm)'
]
group_cols = ['품종코드', '등급코드', '직팜산지코드']

for lag in [1, 2, 3]:
    temp = df[group_cols + ['year', 'week'] + climate_cols].copy()
    temp['week'] = temp['week'] + lag
    temp = temp.rename({col: f"{col}_t-{lag}" for col in climate_cols}, axis=1)
    df = pd.merge(
        df,
        temp,
        on=group_cols + ['year', 'week'],
        how='left'
    )
```

```python
df_corr = df.corr()
df_corr
```
