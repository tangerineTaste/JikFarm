```python
import numpy as np
from sklearn.preprocessing import RobustScaler

df_scaled = df.copy()

df_scaled['ì´ê±°ë˜ëŸ‰(kg)'] = np.log1p(df_scaled['ì´ê±°ë˜ëŸ‰(kg)'])

target = 'í‰ê· ë‹¨ê°€(ì›)'
exclude_cols = ['year', 'week', 'í’ˆì¢…ì½”ë“œ', 'ë“±ê¸‰ì½”ë“œ', 'ì§íŒœì‚°ì§€ì½”ë“œ', target]

numerical_cols = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
scale_cols = [col for col in numerical_cols if col not in exclude_cols]

scaler = RobustScaler()
df_scaled[scale_cols] = scaler.fit_transform(df_scaled[scale_cols])
```

```python
onehot_cols = ['í’ˆì¢…ì½”ë“œ', 'ë“±ê¸‰ì½”ë“œ', 'ì§íŒœì‚°ì§€ì½”ë“œ']
df_encoded = pd.get_dummies(df_scaled, columns=onehot_cols, dtype=int)
```
