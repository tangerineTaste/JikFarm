```python
# ë„ë§¤ì‹œì¥ ì½”ë“œ ë¶ˆëŸ¬ì˜¤ê¸°
df_market = pd.read_csv('ë„ë§¤ì‹œì¥_ì½”ë“œ.csv', encoding='cp949')
```

```python
# í’ˆëª© ì½”ë“œ ì„¤ì •
ITEM_CODES = {
    "ë°°ì¶”": "1001",  # ìš©ê³¤
}

# ë‚ ì§œ ì…ë ¥ (YYYY-MM-DD)  # í…ŒìŠ¤íŠ¸ í›„ì— ì¼ì ì¡°ì • =>
start_date = '2018-01-03'
end_date = '2020-12-31'
start_dt = datetime.strptime(start_date, '%Y-%m-%d')
end_dt = datetime.strptime(end_date, '%Y-%m-%d')
total_days = (end_dt - start_dt).days + 1

# ê¸°íƒ€ ì„¤ì •
max_retries = 3
FAIL_LOG = []
```

```python
# ë””ë ‰í† ë¦¬ ì¤€ë¹„
os.makedirs("logs", exist_ok=True)
os.makedirs("data", exist_ok=True)
```

```python
# í’ˆëª©ë³„ ë°˜ë³µ
for item_name, code in tqdm(ITEM_CODES.items(), desc="ì „ì²´ í’ˆëª© ì§„í–‰"):
    LARGE = code[:2]
    MID = code[2:]
    data_list = []

    print(f"\nğŸ“¦ {item_name} ìˆ˜ì§‘ ì‹œì‘: {start_date} ~ {end_date}")
    print("ğŸ“ˆ ì§„í–‰ë¥ : ", end='')

    current_dt = start_dt
    count = 0

    while current_dt <= end_dt:
        date_str = current_dt.strftime('%Y-%m-%d')  # API í¬ë§·
        for mcode, market_name in df_market.values:
            retry_count = 0
            market_success = False
            page_no = 1

            while retry_count < max_retries:
                try:
                    while True:
                        params = {
                            'serviceKey': API_KEY,
                            'pageNo': page_no,
                            'numOfRows': 100,
                            'cond[trd_clcln_ymd::EQ]': date_str,
                            'cond[whsl_mrkt_cd::EQ]': mcode,
                            'cond[gds_lclsf_cd::EQ]': LARGE,
                            'cond[gds_mclsf_cd::EQ]': MID
                        }

                        response = requests.get(BASE_URL, params=params, verify=False, timeout=10)
                        if response.status_code != 200:
                            fail_reason = f"HTTP {response.status_code}: {response.text[:100]}"
                            retry_count += 1
                            break

                        json_data = response.json()
                        header = json_data.get('response', {}).get('header', {})
                        body = json_data.get('response', {}).get('body', {})
                        items = body.get('items', {}).get('item', [])
                        total_count = int(body.get('totalCount', 0))

                        if isinstance(items, list) and items:
                            data_list.extend(items)
                            market_success = True
                        elif isinstance(items, dict):  # ë‹¨ì¼ ê°ì²´ì¼ ê²½ìš°
                            data_list.append(items)
                            market_success = True
                        else:
                            break  # ë°ì´í„° ì—†ìŒ

                        if page_no * 100 >= total_count:
                            break
                        else:
                            page_no += 1
                            time.sleep(0.3)

                    break  # ë‚´ë¶€ í˜ì´ì§€ ë°˜ë³µ ì„±ê³µ ì‹œ íƒˆì¶œ

                except Exception as e:
                    retry_count += 1
                    fail_reason = f"Exception: {str(e)}"
                    print(f"\nâš ï¸ ì‹¤íŒ¨: {item_name} | {market_name} | {date_str} | ì¬ì‹œë„ {retry_count}íšŒ - {fail_reason}")

                    # ë¡œê·¸ íŒŒì¼ ì €ì¥
                    log_dir = "logs"
                    log_prefix = f"{log_dir}/failed_{item_name}_{mcode}_{date_str}_try{retry_count}"

                    # ì‘ë‹µ ë‚´ìš© ì €ì¥ (ê°€ëŠ¥í•œ ê²½ìš°)
                    if 'response' in locals():
                        content_type = response.headers.get("Content-Type", "")
                        if "text/html" in content_type or response.text.strip().startswith("<"):
                            ext = "html"
                        elif "application/json" in content_type:
                            ext = "json"
                        else:
                            ext = "txt"

                        with open(f"{log_prefix}.{ext}", "w", encoding="utf-8") as f:
                            f.write(response.text)

                        # ì˜¤ë¥˜ ìƒíƒœ ì½”ë“œ ë° ë©”ì‹œì§€ ë³„ë„ ì €ì¥
                        with open(f"{log_prefix}_info.txt", "w", encoding="utf-8") as f:
                            f.write(f"[ì—ëŸ¬ ìƒíƒœ ì½”ë“œ] {response.status_code}\n")
                            f.write(f"[ì˜¤ë¥˜ ë©”ì‹œì§€] {str(e)}\n")
                            f.write(f"[ì‘ë‹µ Content-Type] {content_type}\n")

                    else:
                        with open(f"{log_prefix}_info.txt", "w", encoding="utf-8") as f:
                            f.write(f"[ì‘ë‹µ ì—†ìŒ]\n[ì˜¤ë¥˜ ë©”ì‹œì§€] {str(e)}\n")

                    time.sleep(2 * retry_count)

            if not market_success:
                FAIL_LOG.append({
                    "item": item_name,
                    "market": market_name,
                    "mcode": mcode,
                    "date": date_str,
                    "reason": fail_reason if 'fail_reason' in locals() else 'Unknown'
                })

        # ì§„í–‰ë¥  ì¶œë ¥
        count += 1
        if count % 10 == 0:
            percent = int((count / total_days) * 100)
            print(f"{percent}%", end='', flush=True)
        else:
            print('.', end='', flush=True)

        current_dt += timedelta(days=1)
        time.sleep(0.3)

    print(f"\nâœ… {item_name} ì™„ë£Œ: {len(data_list):,}ê±´")

    if data_list:
        df = pd.DataFrame(data_list)
        filename = f"data/ìœ í†µê³µì‚¬_ë„ë§¤ì‹œì¥_{item_name}_{start_date.replace('-', '')}-{end_date.replace('-', '')}.csv"
        df.to_csv(filename, encoding='cp949', index=False)
        print(f"ğŸ’¾ ì €ì¥ë¨: {filename}")
    else:
        print(f"âš ï¸ {item_name}: ìˆ˜ì§‘ëœ ë°ì´í„° ì—†ìŒ")

# ì‹¤íŒ¨ ë¡œê·¸ ì €ì¥
if FAIL_LOG:
    df_fail = pd.DataFrame(FAIL_LOG)
    df_fail.to_csv('data/ìœ í†µê³µì‚¬_fail_log.csv', index=False, encoding='cp949')
    print(f"\nâ— ì‹¤íŒ¨ ìš”ì²­ {len(FAIL_LOG)}ê±´ ê¸°ë¡ë¨: data/ìœ í†µê³µì‚¬_fail_log.csv")
else:
    print("\nğŸ‰ ëª¨ë“  ìˆ˜ì§‘ ì„±ê³µ, ì‹¤íŒ¨ ì—†ìŒ!")
```

```python
import os
import pandas as pd
import requests
import time
from datetime import datetime
from tqdm import tqdm
import ssl
import warnings
import xml.etree.ElementTree as ET
from requests.packages.urllib3.exceptions import InsecureRequestWarning
from dotenv import load_dotenv

# í™˜ê²½ ë° ê²½ê³  ì„¤ì •
load_dotenv()
warnings.filterwarnings('ignore', category=InsecureRequestWarning)
ssl._create_default_https_context = ssl._create_unverified_context

# ìƒìˆ˜
API_KEY = os.getenv("DO_API_KEY")
BASE_URL = 'http://apis.data.go.kr/B552845/katSale/trades'
ITEM_CODES = {"ë°°ì¶”": "1001"}
max_retries = 2  # ì¬ì‹œë„ ìµœëŒ€ íšŸìˆ˜ (1íšŒ ì‹œë„ + 0íšŒ ì¬ì‹œë„)

# ë””ë ‰í† ë¦¬ ì¤€ë¹„
os.makedirs("logs", exist_ok=True)
os.makedirs("data", exist_ok=True)
os.makedirs("success", exist_ok=True)

# ë„ë§¤ì‹œì¥ ì½”ë“œ ë¶ˆëŸ¬ì˜¤ê¸°
df_market = pd.read_csv("ë„ë§¤ì‹œì¥_ì½”ë“œ.csv", encoding="cp949", header=None)
df_market[0] = df_market[0].astype(str)

# ì‹¤íŒ¨ ë¡œê·¸ ë¶ˆëŸ¬ì˜¤ê¸°
fail_df = pd.read_csv("ìœ í†µê³µì‚¬_fail_log.csv", encoding="cp949")
fail_pairs = fail_df[['mcode', 'date']].drop_duplicates()
fail_pairs['mcode'] = fail_pairs['mcode'].astype(str)

for item_name, code in ITEM_CODES.items():
    LARGE = code[:2]
    MID = code[2:]
    data_list = []

    print(f"\nğŸ“¦ ì‹¤íŒ¨ í•­ëª© ì¬ì‹œë„ ì‹œì‘: {item_name}")
    for _, row in tqdm(fail_pairs.iterrows(), total=len(fail_pairs), desc="ì¬ì‹œë„ ì§„í–‰"):
        mcode = str(row['mcode'])
        date_str = row['date']

        market_name_row = df_market[df_market[0] == mcode]
        if market_name_row.empty:
            print(f"âŒ ì‹œì¥ ì½”ë“œ {mcode} ëˆ„ë½ - ìŠ¤í‚µ")
            continue
        market_name = market_name_row.values[0][1]

        retry_count = 0
        market_success = False

        while retry_count < max_retries:
            page_no = 1
            try:
                while True:
                    print(f"â–¶ï¸ ìš”ì²­ ì‹œë„: {item_name} | ì‹œì¥ì½”ë“œ: {mcode} | ë‚ ì§œ: {date_str} | í˜ì´ì§€: {page_no} | ì¬ì‹œë„: {retry_count + 1}")

                    params = {
                        'serviceKey': API_KEY,
                        'pageNo': page_no,
                        'numOfRows': 100,
                        'cond[trd_clcln_ymd::EQ]': date_str,
                        'cond[whsl_mrkt_cd::EQ]': mcode,
                        'cond[gds_lclsf_cd::EQ]': LARGE,
                        'cond[gds_mclsf_cd::EQ]': MID
                    }

                    response = requests.get(BASE_URL, params=params, verify=False, timeout=10)
                    content_type = response.headers.get("Content-Type", "")
                    print(f"ğŸ“„ ì‘ë‹µ Content-Type: {content_type}, ì‘ë‹µ ê¸¸ì´: {len(response.text)}")

                    time.sleep(2.0)
                    
                    # ì‘ë‹µ í…ìŠ¤íŠ¸ ì•ë¶€ë¶„ í™•ì¸
                    response_preview = response.text[:500].strip()
                    # ì¡°ê±´ ë¶„ê¸° ìƒì„¸í™”
                    if "LIMITED_" in response_preview:
                        fail_reason = "âŒ API í˜¸ì¶œ ì œí•œ (LIMITED_ ì‘ë‹µ)"
                    elif "SERVICE ERROR" in response_preview:
                        fail_reason = "âŒ ì„œë¹„ìŠ¤ ì˜¤ë¥˜ (SERVICE ERROR ì‘ë‹µ)"
                    elif "ERROR" in response_preview.upper():
                        fail_reason = "âŒ ê¸°íƒ€ ì˜¤ë¥˜ í¬í•¨ (ERROR í‚¤ì›Œë“œ í¬í•¨)"
                    elif "TOO MANY REQUESTS" in response_preview.upper():
                        fail_reason = "âŒ ìš”ì²­ ê³¼ë‹¤ë¡œ ì¸í•œ ì œí•œ (Too Many Requests)"
                    else:
                        fail_reason = None  # ì—ëŸ¬ ì•„ë‹˜

                    # ì‹¤ì œ ì—ëŸ¬ ì¡°ê±´ì¼ ê²½ìš°ì—ë§Œ ì¬ì‹œë„ ë£¨í‹´ ì‹¤í–‰
                    if fail_reason:
                        print(f"â›” {fail_reason} - ì¬ì‹œë„ ëŒ€ê¸° ì¤‘ (2ë¶„)")
                        
                        log_prefix = f"logs/retry_failed_{item_name}_{mcode}_{date_str}"
                        with open(f"{log_prefix}.html", "w", encoding="utf-8") as f:
                            f.write(response.text)
                        with open(f"{log_prefix}_info.txt", "w", encoding="utf-8") as f:
                            f.write(f"[ì˜¤ë¥˜] {fail_reason}\n{response_preview}")
                        retry_count += 1
                        if retry_count >= max_retries:
                            print(f"â— ìµœëŒ€ ì¬ì‹œë„ {max_retries}íšŒ ì´ˆê³¼ - ì¤‘ë‹¨")
                            break
                        time.sleep(120)
                        continue

                    if "application/json" in content_type:
                        json_data = response.json()
                        body = json_data.get("response", {}).get("body", {})
                        items = body.get("items", {}).get("item", [])
                        total_count = int(body.get("totalCount", 0))

                    elif "application/xml" in content_type or response.text.strip().startswith("<"):
                        root = ET.fromstring(response.text)
                        total_count_el = root.find(".//totalCount")
                        total_count = int(total_count_el.text) if total_count_el is not None else 0
                        item_els = root.findall(".//item")
                        items = [{el.tag: el.text for el in item} for item in item_els]

                    else:
                        raise ValueError(f"ì•Œ ìˆ˜ ì—†ëŠ” ì‘ë‹µ í˜•ì‹: {content_type}")

                    if not items:
                        print("âš ï¸ ê±°ë˜ ë°ì´í„° ì—†ìŒ - 0ìœ¼ë¡œ ëŒ€ì²´ í›„ ì„±ê³µ ì²˜ë¦¬")
                        items = [0]
                        data_list.extend(items)
                        market_success = True

                        success_log_path = f"success/retry_success_{item_name}_{mcode}_{date_str}_empty.txt"
                        with open(success_log_path, "w", encoding="utf-8") as f:
                            f.write(f"âœ… {datetime.now()} - {item_name} {mcode} {date_str} ê±°ë˜ ì—†ìŒ (ë¹ˆ ì‘ë‹µ)\n")
                        break

                    data_list.extend(items)
                    market_success = True

                    success_log_path = f"success/retry_success_{item_name}_{mcode}_{date_str}.txt"
                    with open(success_log_path, "w", encoding="utf-8") as f:
                        f.write(f"âœ… {datetime.now()} - {item_name} {mcode} {date_str} ë°ì´í„° ìˆ˜ì§‘ ì„±ê³µ\n")

                    if page_no * 100 >= total_count:
                        print(f"âœ… ë§ˆì§€ë§‰ í˜ì´ì§€ ë„ë‹¬ (totalCount: {total_count})")
                        break

                    # ë¬´í•œ ë£¨í”„ ë°©ì§€
                    if page_no > 10:
                        print("ğŸš¨ í˜ì´ì§€ 10 ì´ˆê³¼ - ë¬´í•œ ë£¨í”„ ë°©ì§€ë¥¼ ìœ„í•´ ì¤‘ë‹¨")
                        break

                    page_no += 1
                    time.sleep(1.0)

                if market_success:
                    break  # ì •ìƒì ìœ¼ë¡œ ìˆ˜ì§‘ ì™„ë£Œ ì‹œ retry ë£¨í”„ íƒˆì¶œ
                else:
                    retry_count += 1
                    if retry_count >= max_retries:
                        print(f"â— ìµœëŒ€ ì¬ì‹œë„ {max_retries}íšŒ ì´ˆê³¼ - ì¤‘ë‹¨")
                        break
                    time.sleep(2 * retry_count)

            except Exception as e:
                retry_count += 1
                print(f"â—ì˜ˆì™¸ ë°œìƒ: {e} (ì¬ì‹œë„ {retry_count}/{max_retries})")
                fail_log_prefix = f"logs/retry_failed_{item_name}_{mcode}_{date_str}_try{retry_count}"
                if 'response' in locals():
                    with open(f"{fail_log_prefix}.txt", "w", encoding="utf-8") as f:
                        f.write(response.text)
                with open(f"{fail_log_prefix}_info.txt", "w", encoding="utf-8") as f:
                    f.write(f"[ì˜ˆì™¸] {str(e)}\n")
                if retry_count >= max_retries:
                    print(f"â— ìµœëŒ€ ì¬ì‹œë„ {max_retries}íšŒ ì´ˆê³¼ - ì¤‘ë‹¨")
                    break
                time.sleep(2 * retry_count)

        if not market_success:
            fail_log_path = f"logs/retry_failed_{item_name}_{mcode}_{date_str}.txt"
            with open(fail_log_path, "w", encoding="utf-8") as f:
                f.write(f"âŒ {datetime.now()} - {item_name} {mcode} {date_str} ë°ì´í„° ìˆ˜ì§‘ ì‹¤íŒ¨\n")

    if data_list:
        df = pd.DataFrame(data_list)
        filename = f"data/ìœ í†µê³µì‚¬_retry_{item_name}_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv"
        df.to_csv(filename, encoding='cp949', index=False)
        print(f"âœ… ì €ì¥ ì™„ë£Œ: {filename}")
    else:
        print(f"âš ï¸ {item_name}: ì¬ì‹œë„ì—ì„œë„ ë°ì´í„° ì—†ìŒ")
```
